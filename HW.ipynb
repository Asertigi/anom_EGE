```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def dist(p1, p2):
    return ((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)**0.5
```

```python
fin = open("0.txt")
sz = 100
points = []
eps = 1
clusters = [[]]
mark = [0] * sz

for i in range(sz):
  s=fin.readline()
  x, y = list(map(float, s.split()))
  points.append([x, y])

def dfs(c, i):
    if c == len(clusters):
        clusters.append([points[i]])
    else:
        clusters[c].append(points[i])
    mark[i] = c
    for j in range(sz):
        if mark[j] == 0 and dist(points[i], points[j]) <= eps:
            dfs(c, j)

c = 0
for i in range(sz):
    if mark[i] == 0:
        dfs(c + 1, i)
        c += 1

import random

def generate_color():
    return (random.random(), random.random(), random.random())

i = 0;
flag = 0;

def z(l):
  dist_min=100000000000
  x_min=0
  y_min=0
  i_best=0
  for i in range(len(l)):
    sd=0
    for j in range(len(l)):
      sd += dist(l[i], l[j])
    if (sd < dist_min):
      dist_min=sd
      x_min = l[i][0]
      y_min = l[i][1]
  plt.plot(x_min, y_min, 'b+', markersize=10)
  return [x_min, y_min]
xx=0.0
yy=0.0
count=0
for l in clusters:
    if len(l) >= 30:
        count += 1
        color = generate_color()
        i += 1
        p = z(l)
        xx += p[0]
        yy += p[1]
    else:
        color = 'black'
    x = [point[0] for point in l]
    y = [point[1] for point in l]
    if (color != 'black'):
      plt.scatter(x, y, color=color, label=f'Cluster: {i}')
    elif flag:
      plt.scatter(x, y, color=color)
    else:
      plt.scatter(x, y, color=color, label=f'anomal')
      flag = 1;

plt.legend()
plt.show()

print(xx/count * 100000, yy/count * 100000)



```
